```python
def print_wrapped(number,f,k):
    # Convert the integer to a string
    s = str(number)
    # Loop from 0 to the end of the string, stepping by k
    for i in range(0, len(s), k):
        # Print the slice from current index i to k
        f.write(s[i : i+k])
        f.write("\n")
```
