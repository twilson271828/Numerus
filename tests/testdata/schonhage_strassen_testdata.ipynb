```python
x=ZZ.random_element(10^(k-1),10^k)
y=ZZ.random_element(10^(k-1),10^k)
z=x*y
```

```python
def print_wrapped(number,f,k):
    # Convert the integer to a string
    s = str(number)
    # Loop from 0 to the end of the string, stepping by k
    for i in range(0, len(s), k):
        # Print the slice from current index i to k
        f.write(s[i : i+k])
        f.write("\n")
```

```python
output_file="test1_ss_%d_digits.txt"%k

with open(output_file, 'w') as f:
        f.write(str(x)+"\n")
        f.write(str(y)+"\n")
        f.write(str(z)+"\n")
print("File saved successfully.")
```
